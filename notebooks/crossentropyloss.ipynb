```python
import torch
import torch.nn as nn
from torch import Tensor
import math
```

```python
torch.manual_seed(40)
targets = torch.rand(4)
inputs = torch.rand(4)
targets, inputs
```

```python
loss = nn.CrossEntropyLoss()
```

```python
targets = torch.softmax(targets, dim=-1)
sum(inputs), sum(targets), inputs, targets
```

```python
loss(input=inputs, target=targets)
```

```python
losses = 0

for input, target in zip(inputs, targets):
    losses += math.log(input) * -target

losses
```

```python
a = torch.randn(3, 5, requires_grad=True)
a
```

```python
b = torch.empty(3, dtype=torch.long).random_(5)
b
```

```python
out = loss(a, b)
print(out)
out.backward()
```
